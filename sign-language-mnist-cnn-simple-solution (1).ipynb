***We have a bunch of images representing different ASL gestures. Each image is like a tiny grayscale picture, 28 pixels by 28 pixels in size. We split our data into two CSV files. In each row of these files, the first thing we see is the letter that the gesture represents, like 'A', 'B', 'C', and so on. After that, we have 784 numbers representing the darkness of each pixel in the image.***


# <p style="background-color: black; color: lightblue; padding-right: 100vw; text-align: center; font-family:newtimeroman;padding: 10px; border-radius: 15px 50px; border: 2px solid lightblue;font-size: 33px; text-shadow: 2px 2px 5px rgba(4, 5, 0, 0.5); box-shadow: 5px 5px 15px white;"><span style="color: #e67e22; margin: 0 10px; animation: flicker 1s infinite;">üçÇ</span> About Dataset  üçÉ

- The other 784 columns give us the grayscale values of each pixel in the image.
- There are 24 different letters we're dealing with, from 'A' to 'I' and 'K' to 'Y'.

```python
train = train.drop_duplicates()
test  = test.drop_duplicates()
print(f'{yb_}\nDuplicates has been Removed.{end_}')
```

<span style="font-family:cursive; font-size:36px; padding:8px 16px; border-radius:98px 40px 160px 8px; border-left:8px solid lightblue ; font-weight:bold;"> 7 <span style="color:lighblue ; font-style: italic;"> Labels </span>

```python
label = train['label']
train = train.drop(['label'],axis=1)
test_label = test['label']
test = test.drop(['label'],axis=1)
```

<span style="font-family:cursive; font-size:36px; padding:8px 16px; border-radius:98px 40px 160px 8px; border-left:8px solid lightblue ; font-weight:bold;"> 8 <span style="color:lighblue ; font-style: italic;"> Normalization </span>

```python
train = train /255
test  = test /255
# numpy arrays 
train = np.array(train)
test  = np.array(test)
# 28 by 28 
train = train.reshape(train.shape[0],28,28,1)
test  = test.reshape(test.shape[0],28,28,1)

print(f'{yb_}\n Now Data is Normalized.{end_}')
```

# <p style="background-color: black; color: lightblue; padding-right: 100vw; text-align: center; font-family:newtimeroman;padding: 10px; border-radius: 15px 50px; border: 2px solid lightblue;font-size: 33px; text-shadow: 2px 2px 5px rgba(4, 5, 0, 0.5); box-shadow: 5px 5px 15px white;"><span style="color: #e67e22; margin: 0 10px; animation: flicker 1s infinite;">üçÇ</span> Modelling || CNN Building  üçÉ



<span style="font-family:cursive; font-size:36px; padding:8px 16px; border-radius:98px 40px 160px 8px; border-left:8px solid lightblue ; font-weight:bold;">  <span style="color:lighblue ; font-style: italic;">CNN Architecture </span>

```python
model = models.Sequential()
model.add(layers.Conv2D(64,(3,3),padding ='Same',activation = 'relu',input_shape=(28,28,1)))
model.add(Dropout(0.1))
model.add(BatchNormalization())
model.add(layers.MaxPooling2D(2,2))
model.add(Dropout(0.3))
model.add(layers.Conv2D(128,(3,3),padding ='same',activation='relu'))
model.add(BatchNormalization())
model.add(layers.MaxPooling2D(2,2))
model.add(Dropout(0.3))
model.add(layers.Conv2D(128,(3,3),padding ='same',activation='relu'))
model.add(BatchNormalization())
model.add(layers.MaxPooling2D(2,2))
```

<span style="font-family:cursive; font-size:36px; padding:8px 16px; border-radius:98px 40px 160px 8px; border-left:8px solid lightblue ; font-weight:bold;">  <span style="color:lighblue ; font-style: italic;">Summary  </span>

```python
model.summary()
```

```python
#adding some layers 
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(layers.Dense(25, activation ='softmax'))
```

<span style="font-family:cursive; font-size:36px; padding:8px 16px; border-radius:98px 40px 160px 8px; border-left:8px solid lightblue ; font-weight:bold;">  <span style="color:lighblue ; font-style: italic;">Complie </span>

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['accuracy'])
```

<span style="font-family:cursive; font-size:36px; padding:8px 16px; border-radius:98px 40px 160px 8px; border-left:8px solid lightblue ; font-weight:bold;">  <span style="color:lighblue ; font-style: italic;">Training  </span>

```python
history = model.fit(train,label,epochs=10,validation_data=(test,test_label))
```

<span style="font-family:cursive; font-size:36px; padding:8px 16px; border-radius:98px 40px 160px 8px; border-left:8px solid lightblue ; font-weight:bold;">  <span style="color:lighblue ; font-style: italic;">Evaluation</span>

```python
model.evaluate(test, test_label)
```
